```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import csv

df = pd.read_csv('data_geocoded.csv')
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
df.shape
```

```python
city_murder_data = pd.read_csv('UCR-1985-2015.csv', error_bad_lines=False)
city_murder_data.head()
```

```python
cols_to_use = ['2002_raw_murder_num','2003_raw_murder_num','2004_raw_murder_num','2005_raw_murder_num',
               '2006_raw_murder_num','2007_raw_murder_num','2008_raw_murder_num','2009_raw_murder_num',
               '2010_raw_murder_num','2011_raw_murder_num','2012_raw_murder_num','city','state_short']

city_murder_data = city_murder_data[cols_to_use]
```

```python
city_murder_data.shape
```

```python
city_murder_data.rename(columns = {'city':'City', 'state_short':'State'}, inplace = True)
```

```python
df = pd.merge(df, city_murder_data, on=['City', 'State'], how='left')
df.head()
```

```python
df.shape
```

```python
import numpy as np

print(np.sort(df['xobsyr_0'].unique()))
```

```python
%%time
from tqdm.notebook import tqdm # used to show progress bar

# only keep the value that matches the year, set all other values to NaN

raw_murder_nums = ['2002_raw_murder_num','2003_raw_murder_num','2004_raw_murder_num','2005_raw_murder_num',
                   '2006_raw_murder_num','2007_raw_murder_num','2008_raw_murder_num','2009_raw_murder_num',
                   '2010_raw_murder_num','2011_raw_murder_num','2012_raw_murder_num',]

for i,row in tqdm(df.iterrows(), total=df.shape[0]):
    year = row['xobsyr_0']
    for i in range(len(raw_murder_nums)):
        if int(raw_murder_nums[i][:4]) != int(year):
            df.set_value(i, raw_murder_nums[i], np.nan)
            
df.head()
```

```python
# condense murder_num feature across years into one column

df['murder_num'] = np.nan

for i in range(len(raw_murder_nums)):
    df['murder_num'] = df['murder_num'].fillna(df[raw_murder_nums[i]])

df.drop(columns=raw_murder_nums, inplace=True) # drop extra rows after condensing murder_num feature
df.head()
```

```python
# group murder data

low = df['murder_num'].min()
high = df['murder_num'].max()
quarter = (high - low) / 4

df['murder_numg'] = pd.cut(df['murder_num'], [low,low+quarter,low+3*quarter,high], labels=[0, 1, 2])
df.drop(columns='murder_num', inplace=True)
df.head()
```
